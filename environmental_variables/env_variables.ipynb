Printing Environmental variables in python

```python
import os
 
# printing environment variables
print(os.environ)
```

printing environmental variables in pretty printed manner

```python
for k, v in os.environ.items():
    print(k,":",v)
```

Getting environmental variables values based on keys

Method 1

```python
print(os.environ["USERNAME"])
```

Method 2

```python
print(os.environ.get("USERNAME"))
```

Set an environmental variables

```python
os.environ["dummy"] = "dummy value"
```

```python
print(os.environ.get("dummy"))
```
